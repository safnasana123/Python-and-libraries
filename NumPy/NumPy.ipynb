# NumPy Array Copy vs View

- A copy creates a new array object with its own data
- changes in the copy do NOT affect the original array

```python
import numpy as np
arr = np.array([1,2,3,4,5])
x = arr.copy()
arr[0] = 42 # the copy will not affected by the original ,bcz its diffrnt
print(x)
```

```python
import numpy as np
arr =np.array ([1,2,3,4,5])
x= arr.view()
arr[0] = 42
print(arr)
print(x)
# it should be affected by the changes made to the original array.
```

```python
import numpy as np
arr = np.array([1,2,3,4,5])
x=arr.view()
x[0]=31
print(arr)
print(x)
```

## NumPy Joining Array

we pass a sequence of array that we wants to the "concatenate()" function,along with the axis.If axis is not explicitly passed.
- It is taken as 0.


```python
# join 2 arrays
import numpy as np
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
arr = np.concatenate ((arr1,arr2))
print(arr)
```

```python
# join  two 2-D arrays along rows(axis=1)

import numpy as np
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5,6],[7,8]])
arr = np.concatenate((arr1,arr2),axis=1)
print(arr)
```

## Joining Arrays Using Stack Function

- stacking is same as concatenation ,the only difference is that stacking is done along a new axis.
- we pass a sequence of arrays that we want to join to the "stack()" method along with the axis.
- If axis is not explicitly passed it is taken as 0.

```python
import numpy as np
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr =np.stack((arr1,arr2),axis=1)
print(arr)
```

### Creating the Filter Array

```python
import numpy as np
arr =np.array([41,42,43,44])
# create an empty list
filter_arr = []
for element in arr:
  # if the element is higher than 42,set the value to true,othewise false
  if element > 42:
    filter_arr.append(True)
  else:
    filter_arr.append(False)
newarr = arr[filter_arr]
print(filter_arr)
print(newarr)
```

```python
import numpy as np
arr=np.array([1,2,3,4,5,6,7])
# create an empty list
filter_arr = []
# go through each element in arr
for element in arr:
  # if the element is completely divisible by 2,set the value to true,otherwise false
  if element % 2 == 0:
    filter_arr.append(True)
  else:
    filter_arr.append(False)
newarr = arr[filter_arr]
print(filter_arr)
print(newarr)
```

```python
# subtraction

import numpy as np
arr1=np.array([10,20,30,40,50,60])
arr2=np.array([20,21,22,23,24,25])
newarr=np.subtract(arr1,arr2)
print(newarr)
```

```python
# Multiplication

import numpy as np
arr1=np.array([10,20,30,40,50,60])
arr2=np.array([20,21,22,23,24,25])
newarr=np.multiply(arr1,arr2)
print(newarr)
```

## NumPy LCM Lowest Common Multiple

```python
import numpy as np
num1 = 4
num2 = 6
x = np.lcm(num1,num2)
print(x)
# 4 nteyum 6 nteyum multiplication table l common aayi varunna lowest number aahn 12
```

## Finding LCM in Arrays

```python
import numpy as np
arr = np.array([3,6,9])
x=np.lcm.reduce(arr)
print(x)
# bcz its the lowest common multiple of all three numbers.
# 3*6 = 18 ,6*3 = 18 , 9*3 = 18
```

```python
import numpy as np
arr = np.arange(1,11)
x=np.lcm.reduce(arr)
print(x)
```

## NumPy GCD Greatest Common Denominator

```python
import numpy as np
num1 = 6
num2 = 9
x = np.gcd(num1,num2)
print(x)
# bcz that is the highest number can be divided by (6/3=2 & 9/3 =3)
```

## Finding GCD in Arrays

```python
import numpy as np
arr = np.array([20,8,32,36,16])
x = np.gcd.reduce(arr)
print(x)
# 4, bcz that is the highest number all values can be divided by.
```

# NumPy Set Operations

What is a set?

A set in mathematics is a collection of unique elements.


## Creates Sets in NumPy

```python
# unique()
# convert following array with repeated elements to a set.

import numpy as np
arr =np.array([1,1,1,2,3,4,5,5,6,7])
x=np.unique(arr)
print(x)
```

## Finding Union

```python
# union1d()
# find intersection of the following 2 set arrays

import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([3,4,5,6])
newarr=np.intersect1d(arr1,arr2,assume_unique=True)
print(newarr)
```

## Finding Difference

```python
import numpy as np
set1 =np.array([1,2,3,4])
set2 =np.array([3,4,5,6])
newarr =np.setdiff1d(set1,set2,assume_unique=True)
print(newarr)
```

## Finding Symmetric Difference

```python
# find the symmetric difference of the set1 and set2

import numpy as np
set1=np.array([1,2,3,4])
set2=np.array([3,4,5,6])
newarr=np.setxor1d(set1,set2,assume_unique=True)
print(newarr)
```
