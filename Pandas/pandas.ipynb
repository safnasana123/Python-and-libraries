# Replace Using Mean,Median , or Mode

- A common way to replace empty cells,is to calculate the mean ,median,or mode value of the column.
- Pandas uses the mean(),median(),and mode() methods to calculate the respective values for a specified column

```python
# Calculate the MEAN , and replace any empty values with it

import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/data_frame.csv")
x = df["Calories"].mean()
df.fillna({"Calories": x},inplace =True)
print(df.to_string)
```

```python
# Calculate the MEDIAN , and replace any empty values with it

import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/data_frame.csv")
x=df["Calories"].median()
df.fillna({"Calories": x},inplace =True)
print(df.to_string)
```

```python
# Calculate the MODE ,and replace any empty values with it

import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/data_frame.csv")
x=df["Calories"].mode()[0]
df.fillna({"Calories": x},inplace = True)
print(df.to_string)
```

## Concat

- Concat() function is used to concatenate 2 or more dataframes.
- Contact can do both vertically(axis=0) as well as horizontal(axis=1)

```python
tcs = {"Id":[101,102,103,104],
       "Name":["Mithun","Dipin","Jose","Rahul"],
       "Age":[25,26,27,29]}

wipro = {"Id":[101,102,103,104],
        "Name":["Dev","Sreenath","Sreerag","Shafeel"],
        "Age":[24,24,27,26]}

tcs_emp = pd.DataFrame(tcs)
wipro_emp = pd.DataFrame(wipro)
pd.concat([tcs_emp,wipro_emp])
```

```python
# ignore index - To get common index after concatenation

pd.concat([tcs_emp,wipro_emp],ignore_index=True)
```

```python
# Keys - To provide seperate keys for each concatenate data frames
data=pd.concat([tcs_emp,wipro_emp],keys=["TCS","WIPRO"])
data
```

```python
# Concatenating vertically
data=pd.concat([tcs_emp,wipro_emp],axis=1)
data
```

```python
# Concatenating a dataframe with a series

salary=pd.Series([65000,75000,15000,25000],name="salary")
pd.concat([tcs_emp,salary],axis=1)
```

## Merge

Merge is a process to merge 2 table horizontally

```python
tcs = {"Id":[101,102,103,104],
       "Name":["Mithun","Dipin","Jose","Rahul"],
       "Age":[25,26,27,29]}
wipro = {"Id":[101,102,103,104],
        "Name":["Dev","Sreenath","Sreerag","Shafeel"],
        "Age":[24,24,27,26]}

tcs_emp = pd.DataFrame(tcs)
wipro_emp = pd.DataFrame(wipro)
pd.concat([tcs_emp,wipro_emp])
tcs_emp
```

```python
wipro_emp=pd.DataFrame(wipro)
wipro_emp
```

```python
pd.merge(tcs_emp,wipro_emp,on="Id")
```

```python
pd.merge(tcs_emp,wipro_emp,on="Id",suffixes=["left","right"])
```

```python
pd.merge(tcs_emp,wipro_emp,on="Id",how="outer") # outer join
```

```python
pd.merge(tcs_emp,wipro_emp,on="Id",how="outer",indicator=True)
```

```python
pd.merge(tcs_emp,wipro_emp,on="Id",how="left")  # left join
```

```python
pd.merge(tcs_emp,wipro_emp,on="Id",how="right")  # right join
```

## Cross tab or contingency table

- In pandas , **pd.crosstab()** is used to compute a cross-tabulation of two or more factors.Its often used for analyzing categorical variables,similar to a contingency table in statistics.

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/survey_crosstab.csv")
df
```

```python
# Create cross-tabulation

ct=pd.crosstab(df["Gender"],df["Preferred_Device"])
print(ct)
```

```python
# Cross-tab of AgeGroup vs Preferred_Device with totals

ct2= pd.crosstab(df["AgeGroup"],df["Preferred_Device"],margins = True,margins_name="Total")
print(ct2)
```
