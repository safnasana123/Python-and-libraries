## Matplotlib Markers

```python
# Mark each point with a circle

ypoints = np.array([3,8,1,10])
plt.plot(ypoints,marker="o")
plt.show()
```

```python
# We can give " * "

ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker = "*")
plt.show()
```

## Format String fmt

You can use also use the shortcut string notation parameter to specify the marker.
- This parameter is also called fmt, and is written with this syntax = Marker/line / color.

```python
# mark each point with a circle

ypoints = np.array([3,8,1,10])

plt.plot(ypoints,"o:r")  # r is red color
plt.show()
```

 - "-"   = solid line
 - ":"   = Dotted line
 - "--"  = Dashed line
 - "-."  = Dashed/dotted line

- r = red
- g = green
- b = blue
- c = cyan
- m = magenta
- y = yellow
- k = black
- w = white

### Marker size

# Matplotlib Labels and Title

```python
x=np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y=np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x,y)
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()
```

## Create a Title for a plot

```python
# use the title() function

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x, y)
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()

```

## Set Font Properties for Title and Labels

```python
# We can use the fontdict parameter in xlabel(), ylabel(),and title() to set font properties for the title and labels.

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1={"family":"serif" , "color":"blue" , "size":20}
font2={"family":"serif" , "color":"darkred" , "size":15}

plt.title("Sports Watch Data" , fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage",fontdict = font2)

plt.plot(x,y)
plt.show()
```

## Position the Title

We can use the loc parameter in the title() to position the title.
- Legal values are : "left" ,"right", and "center". Default value is center.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data",loc="left")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x,y)
plt.show()
```

# Matplotlib Subplots

```python
# we can use subplots() function.

# plot 1 :

x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)  # row,column, 1st  subplot position
plt.plot(x,y)

# plot 2 :
x= np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,2,2) # row,column,2nd subplot position
plt.plot(x,y)
plt.show()
```

## The subplots() Function

The subplots() function takes 3 arguments that describes the layout of the figure.
- The layout is organized in rows and columns,which are represented by the first and second arguments.
- The 3rd argument represents the index of the current plot.


```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 3, 1)     # 1st subplot
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 3, 2)    # 2nd subplot
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 3, 3)    # 3rd subplot
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 3, 4)  # 4 th subplot
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 3, 5)  # 5th subplot
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 3, 6)  # 6th subplot
plt.plot(x,y)
plt.show()

```

# Matplotlib Scatter

The scatter() function plots one dot for each observation.It needs 2 arrays of the same length ,one for the values of the x-axis,and one for values on the y-axis:

```python
# A simple scatter plot

x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y)
plt.show()
```

### Compare Plots

```python
# Draw 2 plots on the same figure

# day one,the age and speed of 13 cars

x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y)

# day two , the age and speed of 15 cars

x=np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)
plt.show()
```

### Colors

```python
# We can set your color for each scatter plot with the color or the c argument:

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = "black")
plt.show()
```

### Color Each Dot

```python
# set our own color of markers

x= np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x,y,c=colors)
plt.show()
```

## ColorMap

- The Matplotlib module has a number of available colormaps.
- A colormap is like a list of colors,where each color has a value that ranges from 0 to  100.

### How to use the ColorMap

You can specify the colormap with the keyword argument **cmap** with the value of the colormap,in this case **"viridis"** which is one of the buit-in colormaps available in Matplotlib.

```python
x= np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x,y,c=colors, cmap="viridis")
plt.colorbar()
plt.show()
```

### Size

```python
x= np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x,y,s=sizes)
plt.show()
```

### Alpha

You can adjust the transparency of the dots with the alpha argument.


```python
# set your own size for the markers:

x= np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x,y,s=sizes,alpha=0.5)
plt.show()
```

## Combine color size and Alpha

- you can combine a colormap with different sizes on the dots.This is best  visualized if the dots are transparent.


```python
# Create random arrays with 100 values for x-points,y-points,colors and sizes

import matplotlib.pyplot as plt
import numpy as np

x= np.random.randint(100,size=(100))
y=np.random.randint(100,size=(100))

colors=np.random.randint(100,size=(100))
sizes=10*np.random.randint(100,size=(100))

plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,cmap="nipy_spectral")
plt.colorbar()
plt.show()
```

# Matplotlib Pie Charts

```python
import matplotlib.pyplot as plt
import numpy as np

y=np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

# Labels

```python
# The label parameter must be an array with one label for each wedge

y=np.array([35,25,25,15])
mylabels=["Apples","Banana","Cherries","Dates"]
plt.pie(y,labels=mylabels)
plt.show()
```

## Start Angle

 The startangle parameter is defined with an angle in degrees, default angle is 0

```python
# Start the first wedge at 90 degrees

y=np.array([35,25,25,15])
mylabels=["Apples","Banana","Cherries","Dates"]
plt.pie(y,labels=mylabels,startangle=90)
plt.show()
```

## Explode

```python
# The explode parameter, if specified, and not None, must be an array with one value for each wedge.
# Each value represents how far from the center each wedge is displayed:

#  Pull the "Apples" wedge 0.2 from the center of the pie

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```

# Shadow

Add a shadow to the pie chart by setting the shadows parameter to True

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode=[0.2,0,0,0]

plt.pie(y,labels=mylabels,explode=myexplode,shadow=True)
plt.show()
```

# colors

```python
# specify a new color for each wedge

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black","hotpink","b","red"]
plt.pie(y,labels=mylabels,colors=mycolors)
plt.show()
```

# Legend

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y,labels=mylabels)
plt.legend()
plt.show()
```

## Legend with header

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y,labels=mylabels)
plt.legend(title="Four Fruits:")
plt.show()
```

```python
fig = plt.figure(FigureClass = Waffle,
                 rows=5,
                 columns=10,
                 values=[48,46,6],
                 figsize=(5,3))
plt.show()
```

```python
# Values in dict & Auto-sizing

data = {"Democratic":48,"Republican":46,"Libertarian":3}
fig=plt.figure(FigureClass=Waffle,
               rows=5,
               values=data,
               legend={"loc":"upper left","bbox_to_anchor":(1.1,1)})
plt.show()
```

```python
# Title,legend,colors,background color,block color,direction and style

fig =plt.figure(FigureClass=Waffle,
                rows=5,
                values=data,
                colors=["red","black","blue"],
                title={"label":"Vote Percentage in 2016 US Presidential Election","loc":"left"},
                labels=[f"{v}%"for k,v in data.items()],
                legend={"loc":"lower left","bbox_to_anchor":(0,-0.4),"ncol":len(data),"framealpha":0},starting_location="NW",block_arranging_style="snake")
fig.set_facecolor("white")
plt.show()
```

```python
# plot with icons - pictogram chart

data={"Democratic":48,"Republican":46,"Libertarian":3}
fig = plt.figure(FigureClass=Waffle,
                 rows=5,
                 values=data,
                 colors=["blue","black","red"],
                 legend={"loc":"upper left","bbox_to_anchor":(1,1)},
                 icons="child",
                 font_size=12,
                 icon_legend=True)
plt.show()
```
